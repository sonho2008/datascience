```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.normal(size=100).cumsum()
plt.plot(a)
print(a)
plt.show()
```

```python
Q = pd.tseries.offsets.QuarterEnd()
ind = pd.DatetimeIndex(freq=Q,start=pd.Timestamp('1990-03-31'),periods=len(a))
tsa = pd.Series(a, index = ind)
tsa
```

```python
#plt.style.use('ggplot')
plt.style.use('fivethirtyeight')
#plt.style.use('seaborn-whitegrid')
#plt.style.use('bmh')
fig, ax = plt.subplots(figsize=(10,5))
line = ax.plot(tsa, label='some time series')
ax.legend(loc='upper left')
fig.text(0.5,0.5,'bleh')
plt.suptitle('some title')
plt.show()
plt.close()
```
