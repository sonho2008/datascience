## 3. Missing data and outliers:

Missing data needs to be carefully looked at. The important questions are: Is the missing data biased? In which case we cannot use imputation techniques such as filling in with the mean, or even dropping the incomplete data points is problematic. How should we fill in the data if at all, using the mean, median, by regression or by default value?

Outliers also need to be considered judiciously, they may appear naturally from the data distribution, or they may come from inadequate data collecting process.


### Missing features:


```python
n = len(df_train)
num_miss = df_train.isnull().sum(axis=0)
percent_miss = (num_miss / n)
miss_info = pd.concat([num_miss, percent_miss, df_train.dtypes], axis=1, keys=['number_missing', 'percent_missing', 'dtype'])
miss_info = miss_info.sort_values(by=['number_missing'], ascending=False)
miss_info.loc[miss_info['number_missing']>0]
```

In many cases, missing features actually have meaning. For example, according to the data description, a missing *PoolQC* means there's no pool, missing *MiscFeature* means there is no miscellaneous feature, etc.

So how we will deal with missing values is as follows: for categorical variables (*dtype* is *object*), we will leave them as-is, because in our categorical variable encoding process they will be dealt with (encode the missing info just like another category). For numerical variables (*dtype* is *float* or *int*), we should fill in the missing values either with the mean. median, or by regression of some sort. The numerical columns with missing values are *LotFrontage, GarageYrBlt, MasVnrArea*. Looks like these columns can be filled with the median.




```python
#Filling with median is better than mean for skewed data such as MasVnrArea
#miss_numeric_cols = list(miss_info.loc[(miss_info['number_missing']>0) & (miss_info['dtype']!= np.object)].index)
booarr = np.logical_and(miss_info['number_missing']>0, miss_info['dtype']!= np.object)
miss_numeric_cols = list(miss_info.loc[booarr].index)
print('Median:\n', df_train[miss_numeric_cols].median())
print('Mean:\n', df_train[miss_numeric_cols].mean())
for c in miss_numeric_cols:
    df_train[c] = df_train[c].fillna(value = df_train[c].median())

print('\n\nCheck there\'s no more missing values:')    
df_train[miss_numeric_cols].isnull().sum()
```

## 4. Feature engineering:

### Categorical feature encoding
First we will transform categorical variable into a format understandable by a machine-learning model. A categorical feature may have ordered values, such as *PoolQC* which takes values: *Ex (Excellent), Gd (Good), TA (Average/Typical), Fa (Fair), NA (No Pool)*. These ordered values can be encoded as a single integer representing the score in the range: 4,3,2,1,0; or it can be encoded as 5-dimensional one-hot vector (a vector which has one coordinate = 1 and all other coordinates 0's). The one-hot vector encoding is required for unordered categorical features such as *MiscFeature* which takes values: *Elev (Elevator), Gar2 (2nd Garage),...*, etc. where a single integer score do not make sense.

We will use the one-hot vector encoding as default for all categorical variables since it is better in all cases at a small cost of computational speed and memory. The *pandas.get_dummies* function makes it easy:

```python
print('Number of original categorical columns:', len(df_train_cat.columns))
df_train_cat = pd.get_dummies(df_train_cat, dummy_na=True)
print('Number of new columns:', len(df_train_cat.columns))
print('Training set size:', len(df_train_cat))
df_train_cat.head(5)
```


### Feature Transformations:

Usually we need to transform features to ensure the **LINE** conditition before the final layer of linear regression is performed. **LINE** stands for:

* **L**inearity: the dependent variable (SalePrice) is a linear function of the (transformed) features plus an error term.
* **I**ndependence: the error terms are independent from each other.
* **N**ormality: The error terms are normally distributed with mean 0.
* **E**qual variance: The variance of the error term is constant for different values of predictor variables.

There are 2 approaches to feature transformations. The first is by *trial-and-error*, this takes a lot of manual work and time consuming, but the result (if done well) is more elegant and we gain intuitions about the data in the process. The second approach is having a *neural-network* learn the necessary non-linear functions by which to transform the features. This approach is much more scalable and automated, but with the disadvantage of being a blackbox, and requires more compute power. Careful choices of model evaluation, train/validation/test split, and optimizing process for regularization rate and other model parameters are required to build a good model from the second approach. ($\leftarrow$ Good data science goes here.)

#### Transform the dependent variable:
*SalePrice* is definitely not normally distributed, and as we can see in the scatter plots below, its variance usually increases as the predictor variables increases, violating the **E** in **LINE**. The solution to this is usually to transform the dependent variable by a concave down function like $\log$.

```python
fig = sns.distplot(df_train['SalePrice'], fit = norm)
plt.show()
```

#### Transforming predictor variables:
We favor using a neural-network to learn whatever ways of transforming predictor variables, however we can help it along by providing the most commonly used transformation: the $\log$ function.

Below is an example of why $log$-transforming a predictor variable is a good idea. Let's look at *LotArea* and the scatter plots between it and *logSalePrice* before and after transformation. The plot goes from an ugly mess with outliers (left) into something reasonable (middle). After the  $\log$ transform, the outlier problem goes away as well: dropping the originial outliers (right) does not make much additional improvement.

```python
temp = df_train.loc[:, ['LotArea','logSalePrice']]

temp['logLotArea'] = np.log(temp['LotArea'])
fig, a = plt.subplots(1,3,figsize=(15,5))

noOutInd = (scaled_df['LotArea']<=4)
temp2 = temp.loc[noOutInd,['logSalePrice','LotArea', 'logLotArea']]
temp2.columns = ['logSalePrice', 'LotAreaNoOutliers', 'logLotAreaNoOutliers']

temp.plot.scatter(x='LotArea',y='logSalePrice', ax=a[0])
temp.plot.scatter(x='logLotArea',y='logSalePrice', ax=a[1])
temp2.plot.scatter(x = 'logLotAreaNoOutliers', y='logSalePrice', ax=a[2])
print('Number of outliers:', scaled_df.loc[scaled_df['LotArea']>4,['LotArea']].count())
print('\n correlation coefficients for the 3 plots:')
print(np.corrcoef(temp['LotArea'],temp['logSalePrice'])[0][1],'\t',
      np.corrcoef(temp['logLotArea'],temp['logSalePrice'])[0][1],'\t',
      np.corrcoef(temp2['logLotAreaNoOutliers'],temp2['logSalePrice'])[0][1],'\t')
plt.show()
```

The correlation coefficient also improved after $\log$ transforming *LotArea*.

#### Automated feature transformations
We would like to have a systematic way to determine what features should be $\log$ transformed. So, for all the variable that can be $\log$-transformed, we check if the transformation increases the correlation coefficient between it and *SalePrice*. If it does, we create a new column for the transformed variable, the neural network will decide which is more useful and how to use them in regression.

First we get the predictor variables that can be $\log$ transformed, dropping clearly irrelevant columns. We compute log transform of those columns, and then compute and compare the correlation coefficients with *logSalePrice*. The table output below shows correlation coefficients with *logSalePrice* without and with $\log$ transform, and the improvement difference (the last column).

```python
temp = (df_train_num.min(axis=0)>0)
logable = list(temp.loc[temp == True].index.drop(['logSalePrice', 'Id']))
print('Columns that can be log-transformed:', logable)

ycol = 'logSalePrice'
temp = pd.DataFrame(df_train[ycol], columns=[ycol])
for c in logable:
    logc = 'log'+c
    temp[c] = df_train[c]
    temp[logc] = np.log(df_train[c])
corrmat = temp.corr(method = 'pearson')
corrcoefs = [[corrmat.loc[c, ycol],corrmat.loc['log'+c, ycol]] for c in logable]
corrcoefs = pd.DataFrame(corrcoefs, index=logable, columns = ['NoTransform', 'Log'])
corrcoefs['Log - NT'] = corrcoefs['Log'] - corrcoefs['NoTransform']
corrcoefs = corrcoefs.sort_values('Log - NT', ascending=False)
corrcoefs
```

Considering the scatterplot below, transforming *LotFrontage* pretty much solves the outliers issue, and it improved the correlation coefficient by $0.01$, we will use *Log-NT* $\ge 0.01$ as a threshhold for including the log-transformed feature column into our regression model.

```python
fig, a = plt.subplots(1,2,figsize=(15,5))
temp.plot.scatter('LotFrontage', 'logSalePrice', ax=a[0])
temp.plot.scatter('logLotFrontage', 'logSalePrice', ax=a[1])
plt.show()
```

```python
minvals = df_train_num.min(axis=0)
temp = np.logical_and(minvals > -1, minvals<= 0)
log1pable = list(temp.loc[temp == True].index)
print('Columns that can be log1p transformed:', ', '.join(log1pable),'\nCount:', len(log1pable))
```

```python
temp2 = pd.DataFrame(df_train.loc[:,['TotalBsmtSF', 'logSalePrice']])
temp2['log1pTotalBsmtSF'] = np.log1p(temp2['TotalBsmtSF'])
fig, a = plt.subplots(1,2,figsize=(15,5))
temp2.plot.scatter(y='logSalePrice', x='TotalBsmtSF', ax=a[0])
temp2.plot.scatter(y='logSalePrice', x='log1pTotalBsmtSF', ax=a[1])
plt.show()
```

For the *log1p* transformed variable, a simple neural network of 1 hidden layer with $k \ge 2$ RELU-neurons can learn a piecewise-linear regression line that fits the data. (In general $k$ hidden neurons can learn a $k+1$ piecewise-linear function).
